Question 1. Titanic Data Exploration & Transformation

```python
import numpy as np
import pandas as pd
```

A. Data Loading and Exploration

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
print("first 5 rows: ")
df.head()
```

```python
print("Shape: ", df.shape)
print("Columns: ", df.columns.tolist())
```

```python
print("Missing Values in each column: \n")

print(df.isnull().sum())

```

B. Data Cleaning

```python
# fill missing Age values with the mean age using numpy
df["Age"] = df['Age'].fillna(df['Age'].mean())
df
```

```python
# fill missing Embarked values with the most frequent value
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

```python
# Drop the Cabin column from the dataset
df.drop(columns=['Cabin'], inplace=True)
df
```

C. Feature Engineering

```python
# Create a new column FamilySize = SibSp + Parch + 1.
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
# Convert the Sex column to numerical format using a map.
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
df.head()
```

```python
#Extract the title (e.g., Mr, Mrs, Miss) from the Name column using string methods.
df['Title'] = df['Name'].apply(lambda x: x.split(",")[1].split(".")[0].strip())
df.head()
```

```python
# Create a binary column IsMinor where age < 18 is 1, else 0.
df['IsMinor'] = np.where(df['Age'] < 18, 1, 0)
df.head()
```

D. NumPy Practice

```python
# avg fare of who survived vs not
avg_fare = df.groupby('Survived')['Fare'].mean()
print('Avg Fare: (0 = Not Survived, 1 = Survived):\n',avg_fare)
```

```python
#Std deviation of age
print("Standard deviation of Age: ", np.std(df['Age']))
```

```python
#no. of passengers with age b/w 20 and 40 
count_20_40 = np.sum((df['Age'] >= 20) & (df['Age'] <= 40))
print("Number of passengers aged 20 to 40: ", count_20_40)
```

E. Analysis

```python
#Group the dataset by Pclass and calculate the average survival rate.
survival_rate = df.groupby('Pclass')['Survived'].mean()
print("Surival Rate: \n", survival_rate)
```

```python
#top 3 passengers w highest fare
top3 = df.nlargest(3,'Fare')[['Name','Age','Fare']]
print('Top 3 passengers with highest fare: \n', top3)
```

```python
# cleaned data
df.to_csv("titanic_cleaned.csv", index = False)
```

Question 2. 

student_data.csv

1. Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
student_df = pd.read_csv("student_data.csv")
student_df.head()
```

Linear Regression

```python
x = student_df[['hours_studied', 'attendance']] 
y = student_df['score']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print("Linear Regression Evaluation:")
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
```

```python
# Scatter plot: hours studied vs score
plt.scatter(student_df['hours_studied'], student_df['score'], color='green', label='Actual Score')
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

2. Logistic regression

```python
a = student_df[['hours_studied', 'attendance']]
b = student_df['passed']

a_train, a_test, b_train, b_test = train_test_split(a,b, test_size=0.2, random_state=42)

log_reg = LogisticRegression()
log_reg.fit(a_train, b_train)
```

```python
b_pred = log_reg.predict(a_test)
```

```python
# Evaluation
print("Logistic Regression Evaluation:")
print("Confusion Matrix:\n", confusion_matrix(b_test, b_pred))
print("Accuracy:", accuracy_score(b_test, b_pred))
print("Precision:", precision_score(b_test, b_pred))
print("Recall:", recall_score(b_test, b_pred))
print("F1 score:", f1_score(b_test, b_pred))
```

```python
# Scatter plot: Hours studied vs Pass
plt.scatter(student_df['hours_studied'], student_df['passed'], color='red', alpha=0.6, label='Pass/Fail')
plt.xlabel("Hours Studied")
plt.ylabel("Passed (1=Yes, 0=No)")
plt.title("Hours Studied vs Passing Outcome")
plt.legend()
plt.show()
```
